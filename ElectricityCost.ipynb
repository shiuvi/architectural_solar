```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rc('font',family='Futura')
from IPython.display import display

df = pd.read_csv('APUS49A72610.csv')
df2 = pd.read_csv('APU000072610.csv')
df3 = pd.read_csv('EIA_residentialelectricity.csv')
#Units for df3: revenue (thousand dollars), sales (MWh), customers (count), price (cents/kWh)
```

```python
list(df.columns)
```

```python
date = list(range(2000, 2023, 1))
kwhprice = df['APUS49A72610']
plt.plot(date, kwhprice, color = '#8CC63F')
plt.ylim([0, 0.30])
plt.xlim([1999, 2023])
plt.title('Average Electricity Price per KWh in Los Angeles-Long Beach-Anaheim, CA ', pad=15)
plt.xlabel('Year')
plt.ylabel('$/KWh')
```
